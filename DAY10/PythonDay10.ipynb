```python
import random

# --- Game Explanation ---
# This is a simple number guessing game.
# The computer will choose a random number between 1 and 100.
# Your goal is to guess that number in as few attempts as possible.
# The game will give you hints if your guess is too high or too low.

# 1. Generate a random number
# We use `random.randint(1, 100)` to get a random integer between 1 and 100 (inclusive).
secret_number = random.randint(1, 100)

# 2. Initialize variables
# `attempts` keeps track of how many guesses the player has made.
# `guess` is initialized to 0, or any value that ensures the while loop starts.
attempts = 0
guess = 0

print("Welcome to the Number Guessing Game!")
print("I'm thinking of a number between 1 and 100.")

# 3. Main game loop
# The game continues as long as the player's guess does not match the secret number.
while guess != secret_number:
    try:
        # Get player's input
        # `input()` gets text from the user, `int()` converts it to an integer.
        guess = int(input("Take a guess: "))
        attempts += 1

        # Provide hints
        if guess < secret_number:
            print("Your guess is too low! Try again.")
        elif guess > secret_number:
            print("Your guess is too high! Try again.")
    except ValueError:
        # Handle non-integer input gracefully
        print("Invalid input. Please enter a whole number.")

# 4. End of game
# This code runs once the player has guessed the correct number.
print(f"Congratulations! You guessed the number in {attempts} attempts.")
print(f"The secret number was {secret_number}.")
```
