## Indentation in Python

In Python, **indentation** is not just for readability; it is syntactically significant. Unlike many other programming languages that use curly braces `{}` or keywords like `begin`/`end` to define code blocks, Python uses whitespace (spaces or tabs) to indicate the grouping of statements.

### Key Concepts:
1.  **Code Blocks:** Statements with the same indentation level are considered part of the same code block. These blocks are typically associated with constructs like `if` statements, `for` loops, `while` loops, function definitions (`def`), and class definitions (`class`).
2.  **Consistency:** All statements within a single block *must* have the same indentation level. Mixing tabs and spaces can lead to `IndentationError`.
3.  **Standard:** The Python style guide (PEP 8) recommends using 4 spaces per indentation level.
4.  **Consequences of Incorrect Indentation:**
    *   **`IndentationError`:** If the indentation is inconsistent or incorrect, Python will raise an `IndentationError` or `TabError`.
    *   **Logical Errors:** Even if no error is raised, incorrect indentation can lead to code behaving differently than intended, as statements might be included in or excluded from blocks incorrectly.

### Example:

```python
# This is a function definition.
# The lines inside the function (print and return) are indented by 4 spaces.
def greet(name):
    print(f"Hello, {name}!")
    return f"Greeting for {name} complete."

# This is an if-else statement.
# The print statements within the if and else blocks are indented.
x = 10
if x > 5:
    print("x is greater than 5")
    print("This line is also part of the if block")
else:
    print("x is not greater than 5")

# Calling the function
message = greet("Alice")
print(message)

# Incorrect indentation would lead to an IndentationError:
# def bad_function():
# print("This line is not indented correctly") # This would raise an IndentationError
```
